```python
test_data = get_samples_regression(num_features=100)
print(test_data[0].shape)
print(test_data[1].shape)
```

```python
lr_local = LinearRegression()
lr_local.fit(X_train=test_data[0],Y_train=test_data[1],show_verbose=False,learning_rate=0.001,iterations=1000)
```

```python
lr_local_lasso = LassoRegression()
lr_local_lasso.fit(X_train=test_data[0],Y_train=test_data[1],show_verbose=False,learning_rate=0.001,iterations=1000,penalty_coefficient=0.1)
```

```python
lr_local_ridge = RidgeRegression()
lr_local_ridge.fit(X_train=test_data[0],Y_train=test_data[1],show_verbose=True,learning_rate=0.001,iterations=1000,penalty_coefficient=0.1)
```
