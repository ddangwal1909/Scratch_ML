```python
from sklearn import datasets
from logistic_regression.LogisticRegression import LogisticRegression
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import SGDClassifier as LogisticRegression_sklearn
```

```python
test_data = datasets.load_breast_cancer()
```

```python
X_train, X_test, y_train, y_test = train_test_split(test_data['data'], test_data['target'], test_size=0.10, random_state=234)
```

```python
log_reg_local = LogisticRegression()
log_reg_local.fit(X_train,y_train,learning_rate=0.01,show_verbose=True,iterations=10000)
```

```python
y_preds = log_reg_local.predict(X_test)
print(y_preds.shape,y_test.shape)
log_reg_local.get_accuracy(y_preds,y_test)
```

```python
log_sklearn = LogisticRegression_sklearn(max_iter=10000)
log_sklearn.fit(X_train,y_train)
y_preds_sgd = log_sklearn.predict(X_test)
```

```python
log_reg_local.get_accuracy(y_preds_sgd,y_test)
```
